# Assessing general patterns by group 

```python
# What if we want to create groups in the data frame by a column?
# Let's try the sex column as an example (group females together and group males together)
grouped_surveys = surveys_df.groupby('sex')
```

```python
# Now that we've grouped the data, what is the mean for each column (females vs. males)?
grouped_surveys.mean()
```

# Data visualization 

```python
# We will use Matplot
# Let's first verify matplot is working (if all is well, nothing will show when executed)
%matplotlib inline
```

```python
# Create a summary of counts by subsetting the data
# Count the number of samples for each species
species_counts = surveys_df.groupby('species_id')['record_id'].count()
```

```python
species_counts
```

```python
# Now plot the number of each species
species_counts.plot(kind='bar')
```

```python
# Count the number of unique species in each plot
total_count = surveys_df.groupby('plot_id')['record_id'].nunique()
```

```python
# View the subset data
total_count
```

```python
# Now plot the number of unique species in each plot
total_count.plot(kind='bar')
```

```python
# How would we modify the above code to view the mean weights of each species?
species_weights = surveys_df.groupby('species_id')['weight'].mean()
species_weights.plot(kind='bar')
```

# Creating a data frame and visualization

```python
# Let's make-up a dataset in preparation for creating a stacked bar plot
d = {'one': pd.Series([1,2,3],
         index=['a','b','c']),
    'two': pd.Series([1,2,3,4],
         index=['a','b','c','d'])}
```

```python
# We what this dataset to be viewed as a data frame
my_frame = pd.DataFrame(d)
my_frame.plot(kind='bar', stacked=True, title='Test Stacked Bar Plot') # visualization
```

```python
# Create a stacked bar plot of grouped data by plot and sex (Gender)
# Include the calculated mean weight for each plot
Exercise = surveys_df.groupby(['plot_id','sex'])
Exercise_mean = Exercise['weight'].mean()
```

```python
# Verify the subset data
Exercise_mean.head()                   # What happens if we unstack?
```

```python
# Unstacking creates a column for each group
Unstack = Exercise_mean.unstack()     # Store as a new data frame
```

```python
# Subset data visualization
Unstack.plot(kind='bar', stacked=True, title='Stacked Plot ID by Gender')
```
